## Importing resource and display

```python
original_dataframe  = pd.read_csv('https://raw.githubusercontent.com/IT21034268/shared-location/refs/heads/other-data-01/student-performance-predictions/student_performance-01.csv')
```

```python
edited_dataframe = original_dataframe
```

```python
edited_dataframe.head(5)
```

## Adding a new column 'FinalGradeRemark'

```python
edited_dataframe['FinalGradeRemark'] = np.where( (edited_dataframe['FinalGrade'] >= 0) & (edited_dataframe['FinalGrade'] < 70) , 'Should Improve',
                                  np.where( (edited_dataframe['FinalGrade'] >= 70) & (edited_dataframe['FinalGrade'] <= 100) , 'Performing Better',
                                  "Invalid Mark"))
```

```python
edited_dataframe
```

## Adding column 'NumValue' for which has values from 0 to 8

```python
tempGen = [-99] * ( len(edited_dataframe) )
```

```python
n_columns = len(edited_dataframe.columns)
```

```python
edited_dataframe.insert(( n_columns - 1 ), 'NumValue', tempGen)
```

```python
random_state = np.random.RandomState(seed=13)

for index, row in edited_dataframe.iterrows():
    s = random_state.randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    num_value = np.random.RandomState(seed=s).randint(0, 8)
    edited_dataframe.at[index, 'NumValue'] = ( row['NumValue'] * 0 ) + num_value
```

```python
edited_dataframe
```

## Renaming column 'NumValue' to 'CustomNumValue'

```python
edited_dataframe = edited_dataframe.rename(columns={
                        'NumValue': 'CustomNumValue'
                        })
```

```python
edited_dataframe
```

## Shuffle 'edited_dataframe'

```python
random_state = np.random.RandomState(seed=17)

for index, row in edited_dataframe.iterrows():
    s = random_state.randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    permutation = np.random.RandomState(seed=s).permutation( len(edited_dataframe) )
    edited_dataframe = edited_dataframe.iloc[permutation].reset_index(drop=True)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    s = np.random.RandomState(seed=s).randint(0, 999999)
    permutation = np.random.RandomState(seed=s).permutation( len(edited_dataframe) )
    edited_dataframe = edited_dataframe.iloc[permutation].reset_index(drop=True)
```

```python
edited_dataframe
```

# Delete columns 'StudentID', 'Name', 'CustomNumValue', 'FinalGradeRemark'

```python
columns_to_delete = ['StudentID', 'Name', 'CustomNumValue', 'FinalGradeRemark']
edited_dataframe = edited_dataframe.drop(columns=columns_to_delete)
```

```python
edited_dataframe
```

# Convert 'Gender', 'ParentalSupport' to numeric representation

```python
edited_dataframe['Gender'] = edited_dataframe['Gender'].map({'Male': 1, 'Female': 0})
```

```python
edited_dataframe['ParentalSupport'] = edited_dataframe['ParentalSupport'].map({'Low': 0, 'Medium': 1, 'High': 2})
```

```python
edited_dataframe
```

# Add Normalized Column 'AttendanceRateNormal'

```python
attendance_rate = edited_dataframe.iloc[:, 1:2].values

attendance_rate_normal = (attendance_rate - np.min(attendance_rate)) / (np.max(attendance_rate) - np.min(attendance_rate))
```

```python
edited_dataframe_n_columns = len(edited_dataframe.columns)
```

```python
edited_dataframe.insert(edited_dataframe_n_columns - 5, 'AttendanceRateNormal', attendance_rate_normal)
```

```python
edited_dataframe
```

# Add Normalized Column 'StudyHoursPerWeekNormal'

```python
study_hours_per_week = edited_dataframe.iloc[:, 3:4].values

study_hours_per_week_normal = (study_hours_per_week - np.mean(study_hours_per_week)) / ( np.std(study_hours_per_week))
```

```python
edited_dataframe_n_columns = len(edited_dataframe.columns)
```

```python
edited_dataframe.insert(edited_dataframe_n_columns - 4, 'StudyHoursPerWeekNormal', study_hours_per_week_normal)
```

```python
edited_dataframe
```
